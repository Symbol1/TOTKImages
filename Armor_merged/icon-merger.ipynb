```python
armors = {'BOTW':BOTW_armors, 'TOTK': TOTK_armors}
```

```python
import sys
from PIL import Image

width, height = 256, 256
num_colors = 14
for game in ['BOTW', 'TOTK']:
    for basename in armors[game]:
        print(basename)
        merged_icons = Image.new('RGBA', (width * num_colors, height))
        x_offset = 0
        for color in list_colors:
            colored_name = '../Armor/' + game + '/' + basename + '_' + color + '.png'
            icon = Image.open(colored_name)
            
            merged_icons.paste(icon, (x_offset, 0))
            x_offset += width
            
        merged_icons.save(basename + '_rainbow.png')
```
